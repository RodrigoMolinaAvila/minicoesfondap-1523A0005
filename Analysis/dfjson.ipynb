```python
import pandas as pd
```

```python
df = pd.read_excel("df.xlsx")
```

```python
df.info()
```

```python
df.head()
```

```python
# Paso 1: Reemplazar comas por puntos (coma decimal europea)
df["FechaPublicación"] = df["FechaPublicación"].astype(str).str.replace(",", ".")
df["FechaPublicación2"] = df["FechaPublicación2"].astype(str).str.replace(",", ".")

# Paso 2: Reemplazar 'None' explícitamente y convertir errores a NaN
df["FechaPublicación"] = pd.to_numeric(df["FechaPublicación"], errors="coerce")
df["FechaPublicación2"] = pd.to_numeric(df["FechaPublicación2"], errors="coerce")

# Paso 3: Convertir a datetime desde milisegundos (ignorando nulos)
df["FechaPublicación"] = pd.to_datetime(df["FechaPublicación"], unit="ms", errors="coerce")
df["FechaPublicación2"] = pd.to_datetime(df["FechaPublicación2"], unit="ms", errors="coerce")
```

```python
df.to_excel("df.xlsx", index=False, engine="openpyxl")
```
