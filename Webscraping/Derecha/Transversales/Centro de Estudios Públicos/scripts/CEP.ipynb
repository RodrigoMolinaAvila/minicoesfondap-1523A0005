```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
from concurrent.futures import ThreadPoolExecutor, as_completed

def create_driver(headless=False):
    """Crea y devuelve una instancia del driver de Chrome."""
    options = Options()
    if headless:
        options.add_argument("--headless")
        options.add_argument("--disable-gpu")
    service = Service()
    driver = webdriver.Chrome(service=service, options=options)
    return driver

def extraer_datos_de_url(url, titulo_categoria):
    """Extrae los enlaces, títulos y fechas de una URL y devuelve un DataFrame."""
    driver = create_driver(headless=False)
    driver.get(url)
    print(f"Procesando URL: {url}")
    
    enlaces, titulos, fechas, años = [], [], [], []
    
    try:
        WebDriverWait(driver, 15).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, 'tab'))
        )
        
        botones_anos = driver.find_elements(By.CSS_SELECTOR, '.category-by-year__nav .tab, .tabs .tab')
        
        for boton in botones_anos:
            try:
                boton_texto = boton.text.strip()
                print(f"Clic en el botón: {boton_texto}")
                
                driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", boton)
                time.sleep(1)
                
                try:
                    boton.click()
                except:
                    driver.execute_script("arguments[0].click();", boton)
                
                time.sleep(2)
                
                WebDriverWait(driver, 15).until(
                    EC.presence_of_element_located((By.CLASS_NAME, 'category-by-year__grid'))
                )
                
                secciones = driver.find_elements(By.CLASS_NAME, 'category-by-year__grid')
                
                for seccion in secciones:
                    try:
                        año = seccion.find_element(By.XPATH, './ancestor::section').get_attribute('data-target')
                    except:
                        año = None
                    
                    tarjetas = seccion.find_elements(By.CSS_SELECTOR, 'a.card--category-investigation')
                    
                    for tarjeta in tarjetas:
                        try:
                            enlace = tarjeta.get_attribute('href')
                            titulo = tarjeta.find_element(By.CLASS_NAME, 'card-title').text
                            fecha = tarjeta.find_element(By.CLASS_NAME, 'date').find_element(By.TAG_NAME, 'span').text
                        except:
                            enlace, titulo, fecha = None, None, None
                        
                        enlaces.append(enlace)
                        titulos.append(titulo)
                        fechas.append(fecha)
                        años.append(año)
            except Exception as e:
                print(f"Error al procesar botón {boton_texto}: {e}")
    except Exception as e:
        print(f"Error general en {url}: {e}")
    finally:
        driver.quit()
    
    df = pd.DataFrame({'Enlace': enlaces, 'Título': titulos, 'Fecha': fechas, 'Año': años, 'Producto': titulo_categoria})
    return df

def procesar_enlace(nombre, enlace):
    print(f"Procesando: {nombre} - {enlace}")
    try:
        return extraer_datos_de_url(enlace, nombre)
    except Exception as e:
        print(f"Error en {enlace}: {e}")
        return pd.DataFrame()

dataframes = []

with ThreadPoolExecutor(max_workers=4) as executor:
    futures = [executor.submit(procesar_enlace, row['Nombre'], row['Enlace']) for _, row in pub_ind_links.iterrows()]
    
    for future in as_completed(futures):
        df_categoria = future.result()
        if not df_categoria.empty:
            dataframes.append(df_categoria)

df_final_pi000 = pd.concat(dataframes, ignore_index=True)
df_final_pi000
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

def create_driver(headless=False):
    """Crea y devuelve una instancia del driver de Chrome."""
    options = Options()
    if headless:
        options.add_argument("--headless")
        options.add_argument("--disable-gpu")
    service = Service()
    driver = webdriver.Chrome(service=service, options=options)
    return driver

def extraer_datos(url):
    """Extrae datos navegando por las pestañas y cargando más elementos."""
    driver = create_driver(headless=False)
    driver.get(url)
    
    datos_extraidos = []
    
    try:
        # Esperar a que aparezcan las pestañas
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".tabs button.tab"))
        )

        # Obtener todas las pestañas
        tabs = driver.find_elements(By.CSS_SELECTOR, ".tabs button.tab")

        for tab in tabs:
            tab_text = tab.text.strip()
            print(f"Navegando a pestaña: {tab_text}")

            # Hacer clic en la pestaña
            driver.execute_script("arguments[0].click();", tab)
            time.sleep(2)  # Esperar carga

            # Cargar más datos hasta que el botón esté deshabilitado
            while True:
                try:
                    load_more_button = driver.find_element(By.CSS_SELECTOR, "button.survey-load-more")

                    # Si el botón está deshabilitado, detener el proceso
                    if "disabled" in load_more_button.get_attribute("outerHTML"):
                        print(f"El botón está deshabilitado. No hay más encuestas en {tab_text}.")
                        break

                    # Hacer clic en el botón
                    driver.execute_script("arguments[0].click();", load_more_button)
                    time.sleep(2)  # Esperar nueva carga
                except:
                    print(f"No hay botón 'Cargar más' visible en {tab_text}.")
                    break
            
            # Extraer datos de la tabla
            filas = driver.find_elements(By.CSS_SELECTOR, "section.tab-content.active .table-container table tbody tr")
            for fila in filas:
                try:
                    titulo_elemento = fila.find_element(By.CSS_SELECTOR, "td:nth-child(1) a")
                    titulo = titulo_elemento.text.strip()
                    enlace = titulo_elemento.get_attribute("href")
                    tema = fila.find_element(By.CSS_SELECTOR, "td:nth-child(2) span").text.strip()
                    fecha = fila.find_element(By.CSS_SELECTOR, "td:nth-child(3) div").text.strip()
                    
                    datos_extraidos.append({
                        "Categoría": tab_text,
                        "Título": titulo,
                        "Fecha": fecha,
                        "Tema": tema,
                        "Enlace": enlace
                    })
                except Exception as e:
                    print(f"Error extrayendo datos de una fila: {e}")
                    continue

    except Exception as e:
        print(f"Error general: {e}")
    
    finally:
        driver.quit()
    
    # Convertir datos en DataFrame
    df = pd.DataFrame(datos_extraidos)
    return df

# URL de la página
url = "https://www.cepchile.cl/opinion-publica/encuesta-cep/"
op = extraer_datos(url)

# Mostrar los primeros datos
op
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import pandas as pd
from concurrent.futures import ThreadPoolExecutor, as_completed
import time

def create_driver(headless=True):
    """Inicializa y devuelve un driver de Chrome."""
    options = Options()
    if headless:
        options.add_argument("--headless")
        options.add_argument("--disable-gpu")
    service = Service()
    driver = webdriver.Chrome(service=service, options=options)
    return driver

def es_invalido(valor):
    """Verifica si un valor es inválido."""
    if pd.isna(valor):  
        return True
    if isinstance(valor, str):
        return valor.strip() in ["", "[Vacio]", "[Espacio]"]
    return False

def extract_li_content(driver, url, idx, total):
    """Extrae todos los textos de los elementos <li> dentro del XPath específico."""
    try:
        driver.get(url)

        # Hacer scroll hasta el final de la página para cargar el contenido dinámico
        last_height = driver.execute_script("return document.body.scrollHeight")
        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)  # Esperar para cargar todo el contenido
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height

        # Extraer todos los <li> dentro del XPath
        li_elements = driver.find_elements(By.XPATH, "/html/body/main/article/section[2]/article/section[1]/div/div/div[2]/ul/li")
        
        # Obtener el texto de cada <li>
        texts = [li.text.strip() for li in li_elements]
        
        if not texts:
            texts = None

        print(f"[{idx}/{total}] Extraído contenido de {url[:50]}...")  # Mostrar un resumen de la URL
        return idx, " ".join(texts)  # Unir todos los textos de <li> en una sola cadena
    except Exception as e:
        print(f"Error en {url}: {e}")
        return idx, None

SAVE_FILE = "resultados2.parquet"
# result_df2 = pd.read_parquet(SAVE_FILE)  # Asegúrate de cargar el DataFrame si es necesario
indices_a_reprocesar = result_df2[result_df2["Corpus"].apply(es_invalido)].index
print(f"Se reprocesarán {len(indices_a_reprocesar)} elementos.")

drivers = [create_driver(headless=False) for _ in range(5)]

with ThreadPoolExecutor(max_workers=5) as executor:
    futures = []
    for idx in indices_a_reprocesar:
        row = result_df2.loc[idx]
        driver = drivers[idx % 5]  
        futures.append(executor.submit(extract_li_content, driver, row["Enlace"], idx, len(result_df2)))
    
    for future in as_completed(futures):
        idx, extracted_text = future.result()
        result_df2.at[idx, "Corpus"] = extracted_text

for driver in drivers:
    driver.quit()

result_df2.to_parquet(SAVE_FILE, index=False)
print("Corpus actualizado y guardado.")
```
