```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service

def create_driver(headless=False):
    """Crea y devuelve una instancia del driver de Chrome."""
    options = Options()
    if headless:
        options.add_argument("--headless")
        options.add_argument("--disable-gpu")  # Recomendado en modo headless
    service = Service()  # Se usa el `Service` para evitar advertencias
    driver = webdriver.Chrome(service=service, options=options)
    return driver

driver = create_driver(headless=True)  # Modo headless para mayor eficiencia

# Listas para almacenar los datos
enlaces = []
titulos = []
fechas = []

# Función para extraer datos de una página
def extraer_datos_pagina(driver):
    # Buscar el contenedor de artículos
    contenedor = driver.find_element(By.CSS_SELECTOR, 'body > div.elementor.elementor-78061 > section.elementor-section.elementor-top-section.elementor-element.elementor-element-a7ff49c.elementor-section-boxed.elementor-section-height-default.elementor-section-height-default > div > div > div > div > div > div.elementor-loop-container.elementor-grid')

    # Buscar todos los artículos dentro del contenedor
    articulos = contenedor.find_elements(By.CSS_SELECTOR, 'div[data-elementor-type="loop-item"]')

    # Extraer los datos para cada artículo
    for articulo in articulos:
        try:
            # Enlace
            enlace = articulo.find_element(By.CSS_SELECTOR, 'h1.elementor-heading-title a').get_attribute('href')
            enlaces.append(enlace)
        except:
            enlaces.append(None)
        
        try:
            # Título
            titulo = articulo.find_element(By.CSS_SELECTOR, 'h1.elementor-heading-title a').text
            titulos.append(titulo)
        except:
            titulos.append(None)
        
        try:
            # Fecha
            fecha = articulo.find_element(By.CSS_SELECTOR, 'span.elementor-icon-list-text time').text
            fechas.append(fecha)
        except:
            fechas.append(None)

# URL base de la página a scrapear
base_url = 'https://lyd.org/noticias/'

# Iterar a través de las páginas
pagina_actual = 1
total_paginas = 317  # Número total de páginas

while pagina_actual <= total_paginas:
    # Construir la URL de la página actual
    if pagina_actual == 1:
        url = base_url
    else:
        url = f"{base_url}{pagina_actual}/"

    print(f"Extrayendo datos de la página {pagina_actual}...")

    # Navegar a la página actual
    driver.get(url)

    # Esperar a que se cargue el contenido
    time.sleep(1)  # Ajusta el tiempo de espera según la velocidad de carga

    # Extraer datos de la página actual
    extraer_datos_pagina(driver)

    # Incrementar el número de página
    pagina_actual += 1

# Crear DataFrame
df_not = pd.DataFrame({
    'Enlace': enlaces,
    'Título': titulos,
    'Fecha': fechas
})

# Cerrar el navegador
driver.quit()

# Guardar los datos en un archivo CSV
#df_not.to_csv('noticias_lyd.csv', index=False)

# Mostrar el DataFrame
df_not
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service

def create_driver(headless=False):
    """Crea y devuelve una instancia del driver de Chrome."""
    options = Options()
    if headless:
        options.add_argument("--headless")
        options.add_argument("--disable-gpu")  # Recomendado en modo headless
    service = Service()  # Se usa el `Service` para evitar advertencias
    driver = webdriver.Chrome(service=service, options=options)
    return driver

def scrape_page(driver):
    """Extrae los datos de la página actual."""
    # Espera a que se cargue el contenido (ajustar tiempo según sea necesario)
    time.sleep(1)  # Ajusta el tiempo de espera según la velocidad de carga

    # Buscar el contenedor de los artículos usando el nuevo JS Path
    contenedor_articulos = driver.find_element(By.CSS_SELECTOR, 
        "body > div.elementor.elementor-78379 > section.elementor-section.elementor-top-section.elementor-element.elementor-element-d505c40.elementor-section-boxed.elementor-section-height-default.elementor-section-height-default > div > div > div > div > div > div.elementor-loop-container.elementor-grid")

    # Buscar todos los artículos dentro del contenedor
    articulos = contenedor_articulos.find_elements(By.XPATH, './div')

    # Listas para almacenar los datos
    enlaces = []
    titulos = []
    productos = []  # Agregado el campo de "Producto"
    fechas = []
    autores = []  # Agregado el campo de "Autor"
    medios = []  # Agregado el campo de "Medio"

    # Extraer los datos para cada artículo
    for articulo in articulos:
        try:
            # Enlace: Busca el enlace del artículo
            enlace = articulo.find_element(By.CSS_SELECTOR, 'h1.elementor-heading-title a').get_attribute('href')
            enlaces.append(enlace)
        except:
            enlaces.append(None)
        
        try:
            # Título: Ajusta el selector para el título del artículo
            titulo = articulo.find_element(By.CSS_SELECTOR, 'h1.elementor-heading-title').text
            titulos.append(titulo)
        except:
            titulos.append(None)
        
        try:
            # Producto: Aquí lo tomamos de los términos relacionados, revisando el selector para productos
            producto = articulo.find_element(By.CSS_SELECTOR, '.elementor-post-info__terms-list-item').text
            productos.append(producto)
        except:
            productos.append(None)
        
        try:
            # Fecha: Ajustar el selector para obtener la fecha
            fecha = articulo.find_element(By.CSS_SELECTOR, 'span.elementor-icon-list-text time').text
            fechas.append(fecha)
        except:
            fechas.append(None)
        
        try:
            # Autor: Extraer el autor del artículo
            autor = articulo.find_element(By.CSS_SELECTOR, 'p.elementor-heading-title a[rel="tag"]').text
            autores.append(autor)
        except:
            autores.append(None)
        
        try:
            # Medio: Extraer el medio de publicación
            medio = articulo.find_element(By.CSS_SELECTOR, 'h2.elementor-heading-title').text
            medios.append(medio)
        except:
            medios.append(None)

    # Crear DataFrame con los datos de la página actual
    df_page = pd.DataFrame({
        'Enlace': enlaces,
        'Título': titulos,
        'Producto': productos,  # Añadido el campo Producto
        'Fecha': fechas,
        'Autor': autores,  # Añadido el campo Autor
        'Medio': medios  # Añadido el campo Medio
    })

    return df_page

def main():
    driver = create_driver(headless=False)

    # Lista para almacenar todos los DataFrames de cada página
    all_data = []

    # URL base de la página
    base_url = 'https://lyd.org/opinion-2/'
    page_number = 1
    max_page = 257  # Número máximo de páginas

    while page_number <= max_page:
        # Construir la URL de la página actual
        if page_number == 1:
            url = base_url
        else:
            url = f"{base_url}{page_number}/"

        # Navegar a la página actual
        print(f"Extrayendo datos de la página {page_number}...")
        driver.get(url)

        # Extraer los datos de la página actual
        df_page = scrape_page(driver)

        # Si no hay datos en la página, salir del bucle
        if df_page.empty:
            print(f"No se encontraron datos en la página {page_number}. Deteniendo la extracción.")
            break

        # Agregar los datos de la página actual a la lista
        all_data.append(df_page)

        # Incrementar el número de página
        page_number += 1

    # Cerrar el navegador
    driver.quit()

    # Combinar todos los DataFrames en uno solo
    df_all = pd.concat(all_data, ignore_index=True)

    # Guardar el DataFrame en un archivo CSV (opcional)
    #df_all.to_csv('opiniones_lyd.csv', index=False)

    # Devolver el DataFrame final
    return df_all

if __name__ == "__main__":
    df_op = main()  # Asignar el DataFrame devuelto a una variable
    df_op # Mostrar el DataFrame
```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
from selenium import webdriver
from selenium.webdriver.common.by import By
import threading
import time

# Almacén de WebDrivers por cada hilo
thread_local = threading.local()

def get_driver():
    """Crea o reutiliza un driver para cada hilo."""
    if not hasattr(thread_local, "driver"):
        options = webdriver.ChromeOptions()
        options.add_argument("--headless")  # No mostrar ventana
        options.add_argument("--disable-gpu")  
        options.add_argument("--no-sandbox")  
        options.add_argument("--disable-dev-shm-usage")  
        prefs = {"profile.managed_default_content_settings.images": 2}  # Bloquear imágenes
        options.add_experimental_option("prefs", prefs)
        thread_local.driver = webdriver.Chrome(options=options)
    return thread_local.driver

def extract_corpus(url):
    """Extrae el contenido de una página usando un driver reutilizable."""
    driver = get_driver()
    try:
        driver.get(url)
        time.sleep(1)  # Esperar carga
        paragraphs = driver.find_elements(By.CSS_SELECTOR, 'div.elementor-widget-container p')
        corpus = ' '.join([p.text for p in paragraphs])
    except Exception as e:
        print(f"Error al extraer el corpus de {url}: {e}")
        corpus = None
    return url, corpus

df_copy = lyd.copy()
total_enlaces = len(df_copy)

# Ejecutar en paralelo con 5 workers y reusar los drivers
with ThreadPoolExecutor(max_workers=5) as executor:
    futures = {executor.submit(extract_corpus, url): url for url in df_copy['Enlace']}
    
    for i, future in enumerate(as_completed(futures)):
        url, corpus = future.result()
        df_copy.loc[df_copy['Enlace'] == url, 'Corpus'] = corpus
        print(f"Progreso: {i+1} de {total_enlaces} - Enlace: {url}")

# Cerrar todos los drivers al final
for thread in threading.enumerate():
    if hasattr(thread_local, "driver"):
        thread_local.driver.quit()

df_copy
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import threading
from concurrent.futures import ThreadPoolExecutor

# Almacén de WebDrivers por cada hilo
thread_local = threading.local()

def get_driver():
    """Devuelve un driver único para cada hilo."""
    if not hasattr(thread_local, "driver"):
        options = webdriver.ChromeOptions()
        options.add_argument("--headless")  # No mostrar ventana
        options.add_argument("--disable-gpu")
        options.add_argument("--no-sandbox")
        options.add_argument("--disable-dev-shm-usage")
        prefs = {"profile.managed_default_content_settings.images": 2}  # Bloquear imágenes
        options.add_experimental_option("prefs", prefs)
        thread_local.driver = webdriver.Chrome(options=options)
        print(f"🔧 Nuevo driver inicializado en hilo {threading.get_ident()}")
    return thread_local.driver

def extract_corpus_with_flexible_selectors(url):
    """Extrae el contenido de una página usando diferentes selectores CSS y XPaths."""
    driver = get_driver()
    try:
        print(f"🌐 Accediendo a: {url}")
        driver.get(url)
        time.sleep(1)  # Esperar carga
        
        # Lista de selectores CSS y XPaths a probar
        selectors = [
            {"type": "css", "value": "div.elementor-widget-container p"},
            {"type": "css", "value": "div.page-content p"},
            {"type": "css", "value": "div.cuerpo"},
            {"type": "xpath", "value": "//div[@class='elementor-widget-container']//p"},
            {"type": "xpath", "value": "//div[@class='page-content']//p"},
            {"type": "xpath", "value": "//div[@class='cuerpo']"},
            {"type": "xpath", "value": "//div"}  # Selector amplio para textos en <div>
        ]
        
        corpus = []
        for selector in selectors:
            try:
                if selector["type"] == "css":
                    elements = driver.find_elements(By.CSS_SELECTOR, selector["value"])
                elif selector["type"] == "xpath":
                    elements = driver.find_elements(By.XPATH, selector["value"])
                
                for element in elements:
                    text = element.text.strip()
                    if text:
                        corpus.append(text)
            except Exception as e:
                print(f"⚠️ Error en selector {selector['value']} en {url}: {e}")
        
        corpus_text = " ".join(corpus).strip() if corpus else None
        if not corpus_text:
            print(f"❌ No se pudo extraer texto de {url}")
        
    except Exception as e:
        print(f"❌ Error accediendo a {url}: {e}")
        corpus_text = None
    
    return url, corpus_text

def process_link(index, row, total_enlaces, sub_index, total_sub, df_copy2):
    """Procesa un enlace y extrae su corpus."""
    url = row['Enlace']
    #print(f"🔍 Procesando {index}/{total_enlaces} - {sub_index}/{total_sub}: {url}")
    
    url, corpus = extract_corpus_with_flexible_selectors(url)
    
    if corpus:
        print(f"✅ Corpus extraído correctamente para: {url}")
    else:
        print(f"❌ No se pudo extraer corpus para: {url}")
    
    df_copy2.at[index, 'Corpus'] = corpus  # Evitar problemas de concurrencia

# Copia del DataFrame original
df_copy2 = df_copy.copy()

# Filtrar enlaces a reprocesar
enlaces_a_reprocesar = df_copy2[(df_copy2['Corpus'].isna()) | (df_copy2['Corpus'] == "")]
total_enlaces = len(df_copy2)
total_sub = len(enlaces_a_reprocesar)
print(f"🔄 Total de enlaces a procesar: {total_sub}")

# Procesar enlaces en paralelo con 5 workers
with ThreadPoolExecutor(max_workers=5) as executor:
    for sub_index, (index, row) in enumerate(enlaces_a_reprocesar.iterrows(), 1):
        if row['Tipo'] not in ["Videos", "Podcast", "Encuestas"]:
            executor.submit(process_link, index, row, total_enlaces, sub_index, total_sub, df_copy2)

# Cerrar todos los drivers
if hasattr(thread_local, "driver"):
    thread_local.driver.quit()

print("✅ Proceso completado.")
df_copy2
```
