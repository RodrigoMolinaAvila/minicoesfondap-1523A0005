```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
from bs4 import BeautifulSoup

def obtener_articulos_lacasacomun(url):
    # Configuración de Selenium
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")  # Para no abrir el navegador
    driver = webdriver.Chrome(options=options)

    # Cargar la página
    driver.get(url)

    # Hacer scroll hasta que no carguen más elementos
    last_height = driver.execute_script("return document.body.scrollHeight")
    
    while True:
        # Desplazarse hacia abajo
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Esperar que carguen los nuevos elementos
        
        # Obtener la nueva altura de la página
        new_height = driver.execute_script("return document.body.scrollHeight")
        
        if new_height == last_height:
            break
        last_height = new_height
    
    # Obtener el HTML actualizado
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Buscar todos los artículos
    articles = soup.find_all('div', class_='iSTCpN TBrkhx post-list-item-wrapper blog-post-homepage-description-font blog-post-homepage-description-color blog-post-homepage-description-fill')

    # Listas para almacenar los datos
    titles = []
    links = []
    dates = []
    
    # Extraer la información de cada artículo
    for article in articles:
        # Extraer título
        title = article.find('h2').get_text(strip=True)
        
        # Extraer enlace
        link = article.find('a', href=True)['href']
        
        # Extraer fecha
        date = article.find('span', class_='post-metadata__date').get_text(strip=True)
        
        # Añadir los datos a las listas
        titles.append(title)
        links.append(link)
        dates.append(date)
    
    # Crear un DataFrame
    df = pd.DataFrame({
        'Título': titles,
        'Enlace': links,
        'Fecha': dates
    })

    driver.quit()  # Cerrar el navegador
    return df

# URL del sitio
url = 'https://www.lacasacomun.cl/blog'

# Llamada a la función
df_articulos = obtener_articulos_lacasacomun(url)

# Mostrar el DataFrame
df_articulos
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
from bs4 import BeautifulSoup

def obtener_articulos_lacasacomun(url):
    # Configuración de Selenium
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")  # Para no abrir el navegador
    driver = webdriver.Chrome(options=options)

    # Cargar la página
    driver.get(url)

    # Hacer scroll hasta que no carguen más elementos
    last_height = driver.execute_script("return document.body.scrollHeight")
    scroll_count = 0  # Contador de los desplazamientos (scrolls)
    total_elements = 0  # Contador de los elementos encontrados

    while True:
        # Desplazarse hacia abajo
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Esperar que carguen los nuevos elementos
        
        # Obtener la nueva altura de la página
        new_height = driver.execute_script("return document.body.scrollHeight")
        
        if new_height == last_height:
            break
        last_height = new_height
        
        # Incrementar contador de scrolls
        scroll_count += 1
        
        # Obtener el HTML actualizado y contar los elementos
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        articles = soup.find_all('div', class_='iSTCpN TBrkhx post-list-item-wrapper blog-post-homepage-description-font blog-post-homepage-description-color blog-post-homepage-description-fill')
        
        current_elements = len(articles)
        total_elements = current_elements
        
        # Mostrar en consola cuántos artículos hay
        print(f"Scroll #{scroll_count}: {current_elements} elementos encontrados hasta ahora.")
    
    # Crear un DataFrame con los artículos
    titles = []
    links = []
    dates = []
    
    # Extraer la información de cada artículo
    for article in articles:
        # Extraer título
        title = article.find('h2').get_text(strip=True)
        
        # Extraer enlace
        link = article.find('a', href=True)['href']
        
        # Extraer fecha
        date = article.find('span', class_='post-metadata__date').get_text(strip=True)
        
        # Añadir los datos a las listas
        titles.append(title)
        links.append(link)
        dates.append(date)
    
    # Crear el DataFrame
    df = pd.DataFrame({
        'Título': titles,
        'Enlace': links,
        'Fecha': dates
    })

    # Cerrar el navegador
    driver.quit()
    
    # Mostrar el número total de artículos encontrados
    print(f"Total de artículos encontrados: {len(df)}")
    return df

# URL del sitio
url = 'https://www.lacasacomun.cl/blog'

# Llamada a la función
df_articulos = obtener_articulos_lacasacomun(url)

# Mostrar el DataFrame
df_articulos
```
