```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

# Crear el driver
def create_driver(headless=False):
    options = Options()
    if headless:
        options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    return driver

# Configurar el controlador de Selenium
driver = create_driver()

# Abre el sitio web
driver.get('https://www.cendachile.cl/publicaciones')

# Espera unos segundos para asegurarte que el contenido ha cargado
time.sleep(3)

# Encuentra todos los enlaces dentro del contenedor que mencionas
links = driver.find_elements(By.CSS_SELECTOR, "#WDxLfe > ul > li:nth-child(3) > div.oGuwee.eWDljc.RPRy1e.Mkt3Tc a")

# Extrae los href de los enlaces
link_list = [link.get_attribute('href') for link in links]

# Muestra la lista de enlaces encontrados
print(link_list)

# Cierra el navegador
driver.quit()
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time
from bs4 import BeautifulSoup

# Crear el driver
def create_driver(headless=False):
    options = Options()
    if headless:
        options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    return driver

# Configurar el controlador de Selenium
driver = create_driver()

# Lista para almacenar los resultados de cada pgina
all_paragraphs = []

# Itera sobre cada enlace en la lista
for url in link_list:
    try:
        # Abre la pgina
        driver.get(url)
        
        # Espera un momento para asegurarse de que la pgina est cargada
        time.sleep(3)
        
        # Seleccionamos todo el div que contiene el contenido principal
        content_div = driver.find_element(By.CSS_SELECTOR, "div.hJDwNd-AhqUyc-OwsYgb.Ft7HRd-AhqUyc-OwsYgb.purZT-AhqUyc-II5mzb.ZcASvf-AhqUyc-II5mzb.pSzOP-AhqUyc-qWD73c.Ktthjf-AhqUyc-qWD73c.JNdkSc.SQVYQc")
        
        # Extraemos el texto de todos los prrafos <p> dentro del div
        paragraphs = content_div.find_elements(By.TAG_NAME, 'p')
        paragraph_texts = [p.text for p in paragraphs]
        
        # Agregar los textos extrados de esta pgina a la lista general
        all_paragraphs.extend(paragraph_texts)
        
    except Exception as e:
        print(f"Error al procesar {url}: {e}")

# Crear un DataFrame con los resultados
df = pd.DataFrame(all_paragraphs, columns=['Texto'])

# Cerrar el navegador
driver.quit()

df
```
