# Noticias

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd

def create_driver(headless=False):
    options = Options()
    if headless:
        options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    return driver

driver = create_driver()

# Lista para almacenar los datos
articles_data = []

# Iterar sobre las páginas
for page in range(1, 25):
    url = f"https://chile21.cl/categoria/noticias/page/{page}/"
    driver.get(url)
    time.sleep(1)  # Esperar a que la página cargue

    # Encontrar todos los artículos
    articles = driver.find_elements(By.CSS_SELECTOR, 'article.blog-design-masonry')

    for article in articles:
        try:
            # Extraer el título
            title = article.find_element(By.CSS_SELECTOR, 'h3.wd-entities-title.title.post-title a').text

            # Extraer el autor
            author = article.find_element(By.CSS_SELECTOR, 'a[rel="author"] span.fn').text

            # Extraer la categoría
            category = article.find_element(By.CSS_SELECTOR, 'div.meta-post-categories.wd-post-cat.wd-style-with-bg a').text

            # Extraer el enlace
            link = article.find_element(By.CSS_SELECTOR, 'h3.wd-entities-title.title.post-title a').get_attribute('href')

            # Almacenar los datos
            articles_data.append({
                'Título': title,
                'Autor': author,
                'Producto': category,
                'Enlace': link  # Añadido el enlace
            })
        except Exception as e:
            print(f"Error al extraer datos de un artículo: {e}")

# Cerrar el navegador
driver.quit()

# Crear el DataFrame directamente desde la lista de diccionarios
df_articles = pd.DataFrame(articles_data)

# Mostrar el DataFrame
df_articles
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd

def create_driver(headless=False):
    options = Options()
    if headless:
        options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    return driver

driver = create_driver()

# Lista para almacenar los datos
articles_data = []

url = "https://chile21.cl/publicaciones/"
driver.get(url)
time.sleep(2)  # Esperar a que la página cargue

# Encontrar todos los artículos
articles = driver.find_elements(By.CSS_SELECTOR, 'article.portfolio-entry')

for article in articles:
    try:
        # Extraer el título y el enlace
        title_element = article.find_element(By.CSS_SELECTOR, 'div.wrap-title h3.wd-entities-title a')
        title = title_element.text
        link = title_element.get_attribute('href')
        
        # Extraer la categoría (producto)
        category = article.find_element(By.CSS_SELECTOR, 'div.wrap-meta ul.proj-cats-list li').text
        
        # Almacenar los datos
        articles_data.append({
            'Título': title,
            'Producto': category,
            'Enlace': link
        })
    except Exception as e:
        print(f"Error al extraer datos de un artículo: {e}")

# Cerrar el navegador
driver.quit()

# Crear el DataFrame
df_articles_pub = pd.DataFrame(articles_data)

# Mostrar el DataFrame
df_articles_pub
```

```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException, TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

def create_driver(headless=False):
    """Crea y devuelve una instancia del driver de Chrome."""
    options = Options()
    if headless:
        options.add_argument("--headless")
        options.add_argument("--disable-gpu")  # Recomendado en modo headless
    service = Service()  # Se usa el `Service` para evitar advertencias
    driver = webdriver.Chrome(service=service, options=options)
    return driver

def extraer_corpus_selenium(driver, url):
    """Extrae la fecha, el texto del artículo (corpus) y la fuente desde la URL usando Selenium."""
    try:
        driver.get(url)

        # Esperar hasta que la fecha esté disponible
        try:
            fecha_meta = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'meta[property="article:published_time"]'))
            )
            fecha = fecha_meta.get_attribute('content')
        except TimeoutException:
            fecha = None

        # Esperar hasta que el contenido del artículo esté disponible
        try:
            div_corpus = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'div.wd-text-block'))
            )
            parrafos = div_corpus.find_elements(By.TAG_NAME, 'p')
            corpus = ' '.join([p.text for p in parrafos]).strip() if parrafos else None
        except TimeoutException:
            corpus = None

        # Extraer la fuente
        fuente = None
        try:
            enlace = driver.find_element(By.XPATH, "//a[contains(text(), 'Ver en')]")
            fuente_texto = enlace.text.strip()
            fuente = fuente_texto.split("Ver en", 1)[1].strip() if "Ver en" in fuente_texto else None
        except NoSuchElementException:
            try:
                enlace = driver.find_element(By.CSS_SELECTOR, "a.btn")
                fuente = enlace.text.strip() if enlace else None
            except NoSuchElementException:
                fuente = None

        return fecha, corpus, fuente

    except Exception as e:
        print(f"Error al acceder a {url}: {e}")
        return None, None, None

def actualizar_corpus_selenium(df):
    """Actualiza las columnas 'Fecha', 'Corpus' y 'Fuente' en el DataFrame si están vacías."""
    driver = create_driver(headless=False)  # Se crea el driver en esta función

    try:
        for index, row in df.iterrows():
            if pd.isna(row['Corpus']) or pd.isna(row['Fecha']) or pd.isna(row['Fuente']):
                enlace = row['Enlace']
                fecha, corpus, fuente = extraer_corpus_selenium(driver, enlace)

                if corpus:
                    df.at[index, 'Corpus'] = corpus
                    df.at[index, 'Fecha'] = fecha
                    df.at[index, 'Fuente'] = fuente
    finally:
        driver.quit()  # Cierra el navegador correctamente al finalizar

    return df
```

```python
def extraer_corpus_selenium(driver, url):
    """Extrae la fecha, el texto del artículo (corpus) y la fuente desde la URL usando Selenium."""
    try:
        driver.get(url)

        # Esperar hasta que la fecha esté disponible
        try:
            fecha_meta = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'meta[property="article:published_time"]'))
            )
            fecha = fecha_meta.get_attribute('content')
        except TimeoutException:
            fecha = None

        # Esperar hasta que el contenido del artículo esté disponible
        try:
            div_corpus = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'div.wd-text-block'))
            )
            parrafos = div_corpus.find_elements(By.TAG_NAME, 'p')
            corpus = ' '.join([p.text for p in parrafos]).strip() if parrafos else None
        except TimeoutException:
            corpus = None

        # Extraer la fuente
        fuente = None
        try:
            # Buscar el enlace que contiene "Ver en" o "NOTA COMPLETA EN"
            enlace = driver.find_element(By.XPATH, "//a[contains(text(), 'Ver en') or contains(text(), 'NOTA COMPLETA EN')]")
            fuente_texto = enlace.get_attribute('outerHTML')  # Obtener el HTML completo del enlace
            # Extraer lo que viene después de "Ver en" o "NOTA COMPLETA EN"
            if fuente_texto:
                if "Ver en" in fuente_texto:
                    start = fuente_texto.find("Ver en") + len("Ver en")
                elif "NOTA COMPLETA EN" in fuente_texto:
                    start = fuente_texto.find("NOTA COMPLETA EN") + len("NOTA COMPLETA EN")
                fuente = fuente_texto[start:].strip()  # Tomar lo que viene después y limpiar espacios
        except NoSuchElementException:
            fuente = None

        return fecha, corpus, fuente

    except Exception as e:
        print(f"Error al acceder a {url}: {e}")
        return None, None, None
```
