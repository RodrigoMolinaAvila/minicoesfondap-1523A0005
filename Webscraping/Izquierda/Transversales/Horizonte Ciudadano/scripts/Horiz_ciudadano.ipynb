## Publicaciones

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la página principal de publicaciones
url = "https://www.horizonteciudadano.cl/publicaciones"

# Hacer la solicitud HTTP a la página
response = requests.get(url)

# Comprobar que la solicitud fue exitosa
if response.status_code == 200:
    # Parsear el contenido HTML de la página
    soup = BeautifulSoup(response.text, 'html.parser')

    # Buscar todos los artículos dentro del contenedor con el selector adecuado
    articulos = soup.select('#archivo > div > div:nth-child(1) .col-md-4')

    # Crear listas para almacenar los datos
    titulos = []
    enlaces = []
    fechas = []

    # Iterar sobre los artículos encontrados
    for articulo in articulos:
        # Extraer el enlace del artículo
        enlace = articulo.select_one('a')['href']
        
        # Extraer el título del artículo
        titulo = articulo.select_one('h3').text.strip()

        # Extraer la fecha del artículo
        fecha = articulo.select_one('p em').text.strip() if articulo.select_one('p em') else "Fecha no disponible"

        # Agregar los datos a las listas
        titulos.append(titulo)
        enlaces.append(enlace)
        fechas.append(fecha)

    # Crear un DataFrame de pandas con los datos extraídos
    df = pd.DataFrame({
        'Título': titulos,
        'Enlace': enlaces,
        'Fecha': fechas
    })

else:
    print(f"Error al acceder a la página. Código de estado: {response.status_code}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time

# Crear el driver
def create_driver(headless=False):
    options = Options()
    if headless:
        options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    return driver

driver = create_driver()

# URL base de las publicaciones
base_url = "https://www.horizonteciudadano.cl/publicaciones?page="

# Listas para almacenar los datos
titulos = []
enlaces = []
fechas = []

# Número de página inicial
page_number = 1

while True:
    # Construir la URL de la página actual
    url = base_url + str(page_number)
    
    # Acceder a la página
    driver.get(url)
    
    # Esperar a que el contenido se cargue (puedes ajustar el tiempo si es necesario)
    time.sleep(2)
    
    # Extraer el HTML de la página
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # Buscar los artículos en la página
    articulos = soup.select('#archivo > div > div:nth-child(1) .col-md-4')
    
    # Si no hay artículos, terminar el bucle
    if not articulos:
        print(f"No se encontraron artículos en la página {page_number}. Terminando el scraping.")
        break
    
    # Iterar sobre los artículos encontrados
    for articulo in articulos:
        # Extraer el enlace del artículo
        enlace = articulo.select_one('a')['href']
        
        # Extraer el título del artículo
        titulo = articulo.select_one('h3').text.strip()

        # Extraer la fecha del artículo
        fecha = articulo.select_one('p em').text.strip() if articulo.select_one('p em') else "Fecha no disponible"

        # Agregar los datos a las listas
        titulos.append(titulo)
        enlaces.append(enlace)
        fechas.append(fecha)
    
    # Pasar a la siguiente página
    page_number += 1

# Crear un DataFrame de pandas con los datos extraídos
df = pd.DataFrame({
    'Título': titulos,
    'Enlace': enlaces,
    'Fecha': fechas
})

df

# Cerrar el navegador
driver.quit()
```

## Novedades

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time

# Crear el driver
def create_driver(headless=False):
    options = Options()
    if headless:
        options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    return driver

driver = create_driver()

# URL base de las publicaciones
base_url = "https://www.horizonteciudadano.cl/novedades?page="

# Listas para almacenar los datos
titulos = []
descripciones = []
enlaces = []

# Número de página inicial
page_number = 1

while True:
    # Construir la URL de la página actual
    url = base_url + str(page_number)
    
    # Acceder a la página
    driver.get(url)
    
    # Esperar a que el contenido se cargue (puedes ajustar el tiempo si es necesario)
    time.sleep(2)
    
    # Extraer el HTML de la página
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # Buscar los artículos dentro del contenedor
    articulos = soup.select('body > main > section.mb-5 > div > div.row.d-flex.justify-content-start.g-0 > div.col-md-8.bg-light.px-3.px-md-5.py-3.mb-5')
    
    # Si no hay artículos, terminar el bucle
    if not articulos:
        print(f"No se encontraron artículos en la página {page_number}. Terminando el scraping.")
        break
    
    # Iterar sobre los artículos encontrados
    for articulo in articulos:
        # Extraer el enlace del artículo
        enlace = articulo.select_one('a')['href']
        
        # Extraer el título del artículo
        titulo = articulo.select_one('h3').text.strip()

        # Extraer la descripción (todo el texto dentro de <p>)
        descripcion = articulo.select_one('p').text.strip() if articulo.select_one('p') else "Descripción no disponible"

        # Agregar los datos a las listas
        titulos.append(titulo)
        descripciones.append(descripcion)
        enlaces.append(enlace)
    
    # Pasar a la siguiente página
    page_number += 1

# Crear un DataFrame de pandas con los datos extraídos
nov = pd.DataFrame({
    'Título': titulos,
    'Descripción': descripciones,
    'Enlace': enlaces
})

# Cerrar el navegador
driver.quit()

nov
```

```python
# Volver a crear el driver para extraer la fecha de cada artículo
driver = create_driver()

# Lista para almacenar las fechas
fechas = []

# Recorrer cada enlace para obtener la fecha
total_articulos = len(nov)  # Total de artículos en el DataFrame
for idx, enlace in enumerate(nov['Enlace']):
    driver.get(enlace)
    time.sleep(2)  # Esperar a que la página se cargue
    
    # Extraer el HTML de la página
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # Buscar el contenedor que tiene la fecha
    fecha_elemento = soup.select_one('div.d-flex.justify-content-between > p')
    
    # Extraer la fecha si se encuentra, si no, asignar "Fecha no disponible"
    if fecha_elemento:
        fecha = fecha_elemento.text.strip()
    else:
        fecha = "Fecha no disponible"
    
    # Agregar la fecha a la lista
    fechas.append(fecha)
    
    # Imprimir el progreso
    print(f"Procesando artículo {idx + 1} de {total_articulos}...")

# Cerrar el navegador
driver.quit()

nov2 = nov.copy()

# Agregar la fecha al DataFrame
nov2['Fecha'] = fechas

# Mostrar el DataFrame final
nov2
```

# Concatenación

```python
df_concatenado = pd.concat([df, nov2], ignore_index=True)
df_concatenado
```
