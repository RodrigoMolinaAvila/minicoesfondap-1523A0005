```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import time

# Crear el driver
def create_driver(headless=False):
    options = Options()
    if headless:
        options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    return driver

# Configurar el controlador de Selenium
driver = create_driver()

# URL base de la página
base_url = "https://fundacionsol.cl/blog"

# Lista para almacenar todos los artículos
all_articles = []

# Accedemos a la página inicial
driver.get(base_url)
time.sleep(3)  # Asegúrate de dar tiempo para que cargue la página

page_number = 1
extracted_count = 0

while True:
    print(f"Extrayendo artículos de la página {page_number}... | Artículos extraídos: {extracted_count}")

    # Extraemos los artículos
    articles = driver.find_elements(By.CSS_SELECTOR, "article.o_wblog_post")

    for article in articles:
        # Título del artículo
        title_tag = article.find_element(By.CSS_SELECTOR, "a.o_blog_post_title")
        title = title_tag.text.strip()
        link = title_tag.get_attribute("href")

        # Fecha del artículo
        try:
            date_tag = article.find_element(By.CSS_SELECTOR, "time")
            date = date_tag.text.strip()
        except:
            date = None

        # Categoría del artículo
        try:
            category_tag = article.find_element(By.CSS_SELECTOR, "a[href*='/blog/']")
            category = category_tag.text.strip()
        except:
            category = None

        # Guardamos el artículo
        all_articles.append({
            "Título": title,
            "Enlace": link,
            "Fecha": date,
            "Categoría": category
        })

        # Incrementamos el contador de artículos extraídos
        extracted_count += 1

    # Intentamos encontrar el enlace a la siguiente página
    try:
        next_page_button = driver.find_element(By.LINK_TEXT, "Siguiente")
        next_page_button.click()
        time.sleep(3)  # Esperamos a que la siguiente página cargue
        page_number += 1
    except:
        # Si no hay botón "Siguiente", hemos llegado al final
        print("No hay más páginas.")
        break

# Cerramos el WebDriver
driver.quit()

# Convertimos todos los resultados en un DataFrame de pandas
fs = pd.DataFrame(all_articles)

# Mostrar el DataFrame
fs
```

```python
import pandas as pd
import re

def replace_illegal_with_tilde(df):
    def replace_string(text):
        if isinstance(text, str):
            # Expresión regular para reemplazar caracteres ilegales y la vocal siguiente por la vocal con tilde
            # La expresión busca un carácter ilegal seguido de una vocal (a, e, i, o, u)
            # y la reemplaza con la vocal con tilde correspondiente.
            text = re.sub(r'[\x00-\x1F\x7F](a)', 'á', text)
            text = re.sub(r'[\x00-\x1F\x7F](e)', 'é', text)
            text = re.sub(r'[\x00-\x1F\x7F](i)', 'í', text)
            text = re.sub(r'[\x00-\x1F\x7F](o)', 'ó', text)
            text = re.sub(r'[\x00-\x1F\x7F](u)', 'ú', text)
        return text

    return df.applymap(replace_string)

# Aplicar la función al DataFrame para reemplazar caracteres ilegales
fs_clean = replace_illegal_with_tilde(fs)

# Guardar el DataFrame limpio en Excel
fs_clean.to_excel("Links - Fundación Sol.xlsx", index=False)
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import concurrent.futures

# Crear el driver para cada hilo (cada worker necesita su propio driver)
def create_driver(headless=False):
    options = Options()
    if headless:
        options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    return driver

# Función para extraer la información de cada URL
def extract_corpus_info(url):
    driver = create_driver()
    driver.get(url)
    
    # Esperar a que la página cargue (ajustar según la página)
    time.sleep(1)
    
    # Intentar extraer el Corpus por la primera forma
    try:
        corpus_element = driver.find_element(By.ID, "o_wblog_post_content")
        corpus_text = corpus_element.text.strip()
    except Exception as e:
        # Si no se encuentra el Corpus en la primera forma, probar la segunda ubicación
        try:
            corpus_element = driver.find_element(By.ID, "HOTWordsTxt")
            corpus_text = corpus_element.text.strip()
        except Exception as e:
            # Si no se encuentra el Corpus en ninguna de las dos ubicaciones, buscar el botón de descarga
            try:
                # Buscar el botón con el texto 'DESCARGAR' y obtener su atributo href
                descargar_button = driver.find_element(By.XPATH, "//a[contains(text(), 'DESCARGAR')]")
                corpus_text = descargar_button.get_attribute("href")  # Obtener el enlace completo del atributo href
            except Exception as e:
                corpus_text = None

    # Extraer Fuente (este selector es común para ambas formas de Corpus)
    try:
        fuente_element = driver.find_element(By.CSS_SELECTOR, "#o_wblog_post_content > div.o_wblog_post_content_field.o_wblog_read_text > p:nth-child(1)")
        fuente_text = fuente_element.text.strip().replace("Nota de prensa publicada en", "").strip()
    except Exception as e:
        fuente_text = None

    # Extraer Autor (común para ambas formas de Corpus, si se encuentra)
    try:
        autor_element = driver.find_element(By.XPATH, "//p[contains(text(), 'Por')]")
        autor_text = autor_element.text.strip().replace("Por", "").strip()
    except Exception as e:
        autor_text = None

    # Extraer Categoría (común para ambas formas de Corpus)
    try:
        categoria_element = driver.find_element(By.CSS_SELECTOR, "#o_wblog_post_content > div.css_editable_mode_hidden.text-muted > div:nth-child(1) > a > b")
        categoria_text = categoria_element.text.strip()
    except Exception as e:
        categoria_text = None

    # Extraer Tags (ahora buscar todos los tags dentro del contenedor .css_editable_mode_hidden.text-muted)
    try:
        tags_elements = driver.find_elements(By.CSS_SELECTOR, ".css_editable_mode_hidden.text-muted a.badge")
        tags_text = [tag.text.strip() for tag in tags_elements]
    except Exception as e:
        tags_text = None
    
    # Cerrar el navegador después de extraer los datos
    driver.quit()
    
    return corpus_text, fuente_text, autor_text, categoria_text, ", ".join(tags_text) if tags_text else ""

# Crear una copia del DataFrame original
fs_copy = fs.copy()

# Obtener el total de enlaces
total_enlaces = len(fs_copy["Enlace"])

# Listas para almacenar los resultados
corpus_list = []
fuente_list = []
autor_list = []
categoria_list = []
tags_list = []

# Función para manejar la ejecución en paralelo
def process_url(i, enlace):
    try:
        print(f"Procesando enlace {i}/{total_enlaces}: {enlace}")
        corpus, fuente, autor, categoria, tags = extract_corpus_info(enlace)
        return corpus, fuente, autor, categoria, tags
    except Exception as e:
        print(f"Error al procesar el enlace {i}/{total_enlaces}: {enlace} -> {e}")
        return None, None, None, None, None

# Usar ThreadPoolExecutor para ejecutar en paralelo con 5 workers
with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:
    # Enumerar enlaces y pasarlos junto con el índice
    results = list(executor.map(lambda p: process_url(*p), enumerate(fs_copy["Enlace"], start=1)))

# Procesar los resultados y llenar las listas
for result in results:
    corpus, fuente, autor, categoria, tags = result
    corpus_list.append(corpus)
    fuente_list.append(fuente)
    autor_list.append(autor)
    categoria_list.append(categoria)
    tags_list.append(tags)

# Agregar las nuevas columnas al DataFrame
fs_copy["Corpus"] = corpus_list
fs_copy["Fuente"] = fuente_list
fs_copy["Autor"] = autor_list
fs_copy["Categoria"] = categoria_list
fs_copy["Tags"] = tags_list

# Ver el DataFrame con los nuevos datos
fs_copy
```
