## Publicaciones

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Configuración de Selenium
driver = webdriver.Chrome()  # Asegúrate de tener el driver de Chrome configurado
driver.get("https://www.opeschile.com/publicaciones-blog")  # Reemplaza con la URL del sitio

# Esperamos un momento para que la página cargue completamente
time.sleep(1.3)

# Lista para almacenar los datos
articulos_data = []

# Función para extraer datos de los artículos
def extraer_articulos():
    # Localizamos todos los artículos en la sección indicada
    articulos = driver.find_elements(By.CSS_SELECTOR, "#collection-5f344e27a4b3a20a211d1f05 > div.Site.loaded > div.Site-inner > div > main > section > section > article")

    for articulo in articulos:
        # Extraemos el título, fecha y link
        try:
            titulo = articulo.find_element(By.CSS_SELECTOR, "a.BlogList-item-title").text
            fecha = articulo.find_element(By.CSS_SELECTOR, "time.Blog-meta-item--date").text
            link = articulo.find_element(By.CSS_SELECTOR, "a.BlogList-item-title").get_attribute("href")
            
            # Agregamos los datos extraídos a la lista
            articulos_data.append({
                'Titulo': titulo,
                'Fecha': fecha,
                'Link': link
            })
        except Exception as e:
            print(f"Error al extraer un artículo: {e}")

# Hacemos scroll hasta el final para asegurarnos de que el botón "Más antiguos" aparezca
while True:
    try:
        # Buscamos el botón de "Más antiguos" usando su texto
        cargar_mas_btn = driver.find_element(By.XPATH, "//span[text()='Más antiguos']/parent::a")
        
        # Realizamos scroll hasta el final para cargar el botón "Más antiguos"
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(1)  # Esperamos a que la página cargue más contenido

        # Hacemos clic en el botón para cargar más artículos
        cargar_mas_btn.click()
        time.sleep(1.3)  # Esperamos a que se carguen los artículos
        
        # Extraemos los artículos cargados hasta ahora
        extraer_articulos()
    except:
        # Si no encontramos el botón "Más antiguos", terminamos
        print("No hay más artículos para cargar.")
        break

# Cerramos el navegador
driver.quit()

# Creamos un DataFrame con los datos extraídos
df = pd.DataFrame(articulos_data)

# Mostramos el DataFrame
df
```

```python
df.to_excel("opes_pub.xlsx", index = False)
```

## Actualidad

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Configuración de Selenium
driver = webdriver.Chrome()  # Asegúrate de tener el driver de Chrome configurado
driver.get("https://www.opeschile.com/actualidad")  # URL del sitio

# Esperamos un momento para que la página cargue completamente
time.sleep(3)

# Lista para almacenar los datos
articulos_data = []

# Función para extraer datos de los artículos
def extraer_articulos():
    # Localizamos todos los artículos en la sección indicada
    articulos = driver.find_elements(By.CSS_SELECTOR, "div.summary-item")
    
    print(f"Artículos encontrados: {len(articulos)}")  # Verificación de artículos encontrados

    for articulo in articulos:
        try:
            # Título del artículo
            titulo = articulo.find_element(By.CSS_SELECTOR, "a.summary-title-link").text
            
            # Fecha del artículo
            fecha = ""
            fecha_element = articulo.find_elements(By.CSS_SELECTOR, "time.summary-metadata-item--date")
            if fecha_element:
                fecha = fecha_element[0].get_attribute("datetime")
            
            # Link del artículo
            link = articulo.find_element(By.CSS_SELECTOR, "a.summary-title-link").get_attribute("href")
            
            # Categoría del artículo
            categoria = ""
            categoria_element = articulo.find_elements(By.CSS_SELECTOR, "a.summary-metadata-item--category")
            if categoria_element:
                categoria = categoria_element[0].text

            # Agregamos los datos extraídos a la lista
            articulos_data.append({
                'Titulo': titulo,
                'Fecha': fecha,
                'Link': link
            })
        except Exception as e:
            # En caso de cualquier error, lo mostramos y continuamos con el siguiente artículo
            print(f"Error al extraer un artículo: {e}")
            articulos_data.append({
                'Titulo': "",
                'Fecha': "",
                'Link': ""
            })

# Extraemos los artículos
extraer_articulos()

# Cerramos el navegador
driver.quit()

# Creamos un DataFrame con los datos extraídos
act = pd.DataFrame(articulos_data)

# Mostramos el DataFrame
print(act)
```

# Concatenación

```python
# Concatenar df y act en un único DataFrame llamado opes
opes = pd.concat([df, act], ignore_index=True)

# Mostrar el DataFrame concatenado
opes
```

```python
opes.to_excel("OPES.xlsx", index=False)
```
