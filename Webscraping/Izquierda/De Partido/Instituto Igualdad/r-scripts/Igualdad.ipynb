```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd  # Importar pandas para manejar los datos en formato DataFrame

# Crear el driver
def create_driver(headless=False):
    options = Options()
    if headless:
        options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    return driver

# Función para hacer scroll hasta el botón "Cargar más" y centrarlo
def scroll_to_load_more():
    # Encuentra el botón "Cargar más"
    load_more_button = driver.find_element(By.ID, 'next-page-tdi_69')
    
    # Obtener la altura de la ventana del navegador
    window_height = driver.execute_script("return window.innerHeight")
    
    # Obtener la posición Y del botón "Cargar más"
    button_location = load_more_button.location['y']
    
    # Calcular el desplazamiento necesario para centrar el botón
    scroll_position = button_location - (window_height / 2) + (load_more_button.size['height'] / 2)
    
    # Realizar el scroll hasta la posición calculada
    driver.execute_script(f"window.scrollTo(0, {scroll_position});")
    
    time.sleep(2)  # Espera a que el scroll se complete

# Función para obtener los artículos actuales
def get_articles():
    articles = driver.find_elements(By.CSS_SELECTOR, '.tdb_module_loop')
    article_data = []
    
    for article in articles:
        # Obtener enlace, título y categoría de cada artículo
        link = article.find_element(By.CSS_SELECTOR, '.entry-title a').get_attribute('href')
        title = article.find_element(By.CSS_SELECTOR, '.entry-title a').text
        category = article.find_element(By.CSS_SELECTOR, '.td-post-category').text
        
        # Guardamos los datos
        article_data.append({'link': link, 'title': title, 'category': category})
    
    return article_data

# Lista de URLs y sus nombres correspondientes
urls = {
    "Noticias": "https://institutoigualdad.cl/categoria/noticias/",
    "Igualdad en los Medios": "https://institutoigualdad.cl/categoria/igualdad-en-los-medios/",
    "Opinión": "https://institutoigualdad.cl/categoria/columnas/opinion/",
    "Editorial": "https://institutoigualdad.cl/categoria/editorial/",
    "Columnas": "https://institutoigualdad.cl/categoria/columnas/",
    "Documentos": "https://institutoigualdad.cl/categoria/documentos/",
    "Eventos": "https://institutoigualdad.cl/categoria/eventos/"
}

# Crear el driver
driver = create_driver()

# Lista para almacenar todos los artículos
all_articles = []

# Iterar sobre las URLs y extraer los artículos
for section, url in urls.items():
    driver.get(url)
    
    # Variable para contar el número de páginas que hemos cargado
    n = 0
    while True:
        # Obtener los artículos de la página actual
        articles_data = get_articles()
        
        # Añadir la sección a cada artículo (el nombre de la URL)
        for article in articles_data:
            article['section'] = section
        
        # Añadir los artículos extraídos a la lista
        all_articles.extend(articles_data)
        
        # Verificar si el botón "Cargar más" está presente
        try:
            load_more_button = driver.find_element(By.ID, 'next-page-tdi_69')
            # Hacemos scroll hasta el botón "Cargar más" y lo centramos
            scroll_to_load_more()
            
            # Hacemos clic en el botón de "Cargar más"
            load_more_button.click()
            
            # Esperamos un momento para que carguen más artículos
            time.sleep(1.3)
            n += 1
            print(f"Sección: {section}, Páginas cargadas: {n}")
        except:
            # Si el botón no está presente, significa que hemos llegado a la última página
            print(f"Sección {section}: No hay más artículos para cargar.")
            break

# Crear un DataFrame de Pandas con los datos extraídos
ig = pd.DataFrame(all_articles)

# Mostrar el DataFrame resultante
ig

# ig.to_excel('Ing_igualdad.xlsx', index=False)

# Cerrar el driver cuando termine
driver.quit()
```
